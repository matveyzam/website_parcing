```python
import requests
from bs4 import BeautifulSoup
url = "https://www.basketballnews.com"
res = requests.get(url)
print(res.status_code)
print(res.text)
```

```python
import pandas as pd

data = {
    "title": [],
    "author": [],
    "date": [],
    "a": []
}

soup = BeautifulSoup(res.text, "lxml")
posts = soup.find_all("article", "css-175oi2r")
for post in posts:
  title = post.find("h1", "css-1rynq56")
  title = title.text if title else "unknown"
  author = post.find("div", "css-1rynq56")
  author = author.text.lower().replace("by ", "").title() if author else "unknown"
  date = post.find("div", "css-1rynq56 r-dnmrzs r-1udh08x r-1udbk01 r-3s2u2q r-1iln25a r-1niwhzg r-1az3528 r-1231a37 r-1enofrn r-356f0p r-waer2z r-tsynxw r-1xnzce8")
  date = date.text if date else "unknown"

  a = post.find('a')
  a = "https://www.basketballnews.com"+a.get("href") if a else "unknown"

  if title not in data["title"]:
    data["title"].append(title)
    data["author"].append(author)
    data["date"].append(date)
    data["a"].append(a)

df = pd.DataFrame(data, columns=["title", "author", "date", "a"])
df.to_excel('basketball.xlsx')
df.to_csv('basketball.csv')
df
```
